```python
import cv2 as cv
import glob
import matplotlib.pyplot as plt
import Comparator
```

```python
path = "SomeCorelImages/*" #10 Random Images From Corel Dataset
data = []
files = glob.glob(path)
for file in files:
    image = cv.imread(file)
    data.append(cv.cvtColor(image, cv.COLOR_BGR2RGB))
query = cv.cvtColor(cv.imread('query.jpg'), cv.COLOR_BGR2RGB)
```

```python
C = Comparator.Comparator(data, query)
output1 = C.colorHistogram()
output2 = C.lbp()
```

```python
f = plt.figure(figsize=(15,15))
f.add_subplot(1,3,1)
plt.imshow(query)
plt.title('Input image')
f.add_subplot(1,3,2)
plt.imshow(output1)
plt.title('Closest image by Color Histogram')
f.add_subplot(1,3,3)
plt.imshow(output2)
plt.title('Closest image by LBP')
plt.show()
```
