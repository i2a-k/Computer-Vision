```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
im = cv.cvtColor(cv.imread('City.jpg'), cv.COLOR_BGR2RGB)
obj= cv.cvtColor(cv.imread('Eagle.jpg'), cv.COLOR_BGR2RGB)
obj = cv.resize(obj, (int(obj.shape[1]*2/3),int(obj.shape[0]*2/3)))
```

```python
f = plt.figure()
f.add_subplot(1,2,1)
plt.imshow(im)
plt.title('Background Image')
f.add_subplot(1,2,2)
plt.imshow(obj)
plt.title('Object Image')
plt.show()
```

```python
mask = 255 * np.ones(obj.shape, obj.dtype)
width, height, _ = obj.shape
objLoc = (int(width*5), int(height/2))
clonedImage = cv.seamlessClone(obj, im, mask, objLoc, cv.NORMAL_CLONE)
```

```python
f = plt.figure()
plt.imshow(clonedImage)
plt.title('Cloned Image')
plt.show()
```
